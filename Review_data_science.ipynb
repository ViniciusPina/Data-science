```python
def limites(coluna):
    Q1 = coluna.quantile(0.25) #quantile = quartil
    Q3 =  coluna.quantile(0.75)
    amplitude = Q3- Q1
    limite_inferior = Q1 - 1.5 * amplitude
    limite_superiror =  Q3 + 1.5 * amplitude
    return limite_inferior,limite_superiror


def diagrama_caixa(coluna):
    fig,(ax1,ax2) = plt.subplots(1,2)
    fig.set_size_inches(15,5)
    sns.boxplot(x=coluna,ax=ax1)
    ax2.set_xlim(limites(coluna))
    sns.boxplot(x=coluna,ax=ax2)
    return



diagrama_caixa(vendas3000['Quantidade Vendida'])
```

```python
import pandas as pd

# Supondo que 'vendas3000' seja seu DataFrame com colunas 'Quantidade Vendida', 'Quantidade Devolvida' e 'Categoria'
# Carregue seus dados...

# Criando listas de vendas, devoluções e categorias
lista_vendas = vendas3000['Quantidade Vendida'].tolist()
lista_devolucoes = vendas3000['Quantidade Devolvida'].tolist()
lista_Categoria = vendas3000['Categoria'].tolist()

# Criando um DataFrame a partir das listas
df = pd.DataFrame({
    'Categoria': lista_Categoria,
    'Quantidade Vendida': lista_vendas,
    'Quantidade Devolvida': lista_devolucoes
})

# Groupby pela categoria e somando as vendas e devoluções por categoria
vendas_por_categoria = df.groupby('Categoria').agg({
    'Quantidade Vendida': 'sum',
    'Quantidade Devolvida': 'sum'
}).reset_index()

# Exibindo as vendas totais e devoluções por categoria
display(vendas_por_categoria)




```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que 'vendas_por_categoria' já tenha sido calculado conforme mencionado anteriormente

# Calculando a porcentagem das vendas por categoria
total_vendas = vendas_por_categoria['Quantidade Vendida'].sum()
vendas_por_categoria['Porcentagem'] = vendas_por_categoria['Quantidade Vendida'] / total_vendas * 100



# Calculando a porcentagem de devoluções em relação às vendas
vendas_por_categoria['Porcentagem Devolvida'] = (vendas_por_categoria['Quantidade Devolvida'] / 
                                                 vendas_por_categoria['Quantidade Vendida']) * 100

# Encontrando a categoria com a maior porcentagem
categoria_maior_porcentagem = vendas_por_categoria.loc[vendas_por_categoria['Porcentagem'].idxmax(), 'Categoria']

# Criando a lista 'explode' para destacar a categoria com a maior porcentagem
explode = [0.1 if cat == categoria_maior_porcentagem else 0 for cat in vendas_por_categoria['Categoria']]

# Criando o gráfico de pizza
labels = vendas_por_categoria['Categoria']
quantidade_vendida = vendas_por_categoria['Quantidade Vendida']

plt.figure(figsize=(8, 8))
plt.pie(quantidade_vendida, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.title('Quantidade Vendida por Categoria')

plt.show()
```
